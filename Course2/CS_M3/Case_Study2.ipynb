```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
df = pd.read_csv('cereal.csv')

df.head()
```

```python
df1 = df[['name','sugars','vitamins']]
df1
```

```python
bins = np.linspace(0, 1.0, 40)
plt.figure(figsize=(10,10))
plt.hist(df1[['sugars','vitamins']], alpha = 0.5, label=df1['name'])
plt.legend()
```

```python
new = {'N': 'Nabisco',
'Q': 'Quaker Oats',
'K': 'Kelloggs',
'R': 'Raslston Purina',
'G': 'General Mills' ,
'P' :'Post' ,
'A':'American Home Foods Products'}

for i in range(0,len(df)):
    df.at[i , 'mfrs'] = new[df.iloc[i]['mfr']]
```

```python
sns.countplot(df['mfrs'], orient='v')
plt.xticks(rotation=90)
##unabel to plot horizontally Please suggest in review
## will plot with bar plot
```

```python
new = dict(df.groupby('mfrs').count()['name'])
lt = []
lt1 = []

for i in new.keys():
    lt.append(i)

for i in new.values():
    lt1.append(i)

plt.barh(lt,lt1, color='k')
plt.xlabel('Count of Cereals by manufacturer')
plt.ylabel('Name of Manufacturer')
plt.title('Manufacturer vs Count')
```
