```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('prisoners.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
del df['total_benefitted']
df['new_col'] = df.sum(axis=1)
df['total_benefitted'] = df['new_col'] - df['YEAR']
del df['new_col']
```

```python
df.head()
```

```python
df.tail()
```

```python
new = dict(df.sum(axis=0))

new['STATE/UT'] = "Total Count"
new['YEAR'] = "NA"

df1 = df.append(new, ignore_index=True)
```

```python
df1.tail()
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('State/UT')
plt.ylabel('Count of Total Benefitted')
plt.title('State vs Total Benefitted')
plt.xticks(rotation=90)
plt.bar(df['STATE/UT'], df['total_benefitted'])
```

```python
new = dict(df.sum(axis=0))
new.pop('STATE/UT')
new.pop('YEAR')
new.pop('total_benefitted')
new

plt.figure(figsize=(10,10))
plt.pie(new.values(), labels=new.keys(), autopct='%f2')
```
