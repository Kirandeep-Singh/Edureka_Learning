```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BigMartSalesData.csv')
df.head()
```

```python
df1 = pd.DataFrame()
tmpdf = df[df['Year'] == 2011]
df1['Month'] = tmpdf['Month']
df1['Amount'] = tmpdf['Amount']
tmp = df1.groupby('Month').sum()

dict = tmp.to_dict()['Amount']

x = [i for i in dict.keys()]
y = [i for i in dict.values()]

plt.figure(figsize=(10,10), dpi=200)
fig, ax = plt.subplots(2)
fig.suptitle('Both Graphs Together')
ax[0].bar(x,y)
ax[1].plot(x,y)
```

```python
df2 = pd.DataFrame()

df2['Country'] = tmpdf['Country']
df2['Amount'] = tmpdf['Amount']
tmp = df2.groupby('Country').sum()

country_dict = tmp.to_dict()['Amount']
plt.figure(figsize=(10,10))

plt.pie([float(i) for i in country_dict.values()], labels=(country_dict.keys()), autopct='%0.2f', shadow=True, startangle=90)
```
